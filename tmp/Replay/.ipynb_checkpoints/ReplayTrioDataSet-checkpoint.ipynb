```python
# matplotlib.use('Qt5Agg')

import matplotlib.pyplot as plt
```

```python
import mne
```

```python
raw=mne.io.read_raw_eeglab('trio2_eoec_in-noscan_hpump-off.set')
```

```python
m=raw[:,:][0]
m=numpy.array(m.transpose())
ch=raw.ch_names
fs=raw.info['sfreq']
```

```python
import libmushu
amp = libmushu.get_amp('replayamp')
```

```python
amp.configure(m, (), ch, fs, realtime=True, blocksize_samples=20)
```

```python
amp.start()
```

```python
alld=[]
allm=[]
import time
starttime=time.time()
newtime=starttime
i=0;
while time.time() - starttime < 5:

    while time.time() - newtime < 0.5:
        pass
    else:
        data, marker = amp.get_data()
        alld.append(data)
        for m in marker:
            allm.append(marker)

        print('%d' % i, end='', flush=True)
        # print(data.shape)
        # print(marker)
        i+=1
        newtime+=0.5
```

```python
amp.stop()
```

```python
m=numpy.concatenate(alld)
```

```python
%matplotlib qt5
plt.figure()
plt.plot(m[:10000, 33:38])
```
